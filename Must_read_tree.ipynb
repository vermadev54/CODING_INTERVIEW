```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class BST:
    def insert(self, root: TreeNode, val: int) -> TreeNode:
        # If the tree is empty, create a new node as the root
        if root is None:
            return TreeNode(val)

        # Traverse to the correct position and insert the new node
        if val < root.val:
            root.left = self.insert(root.left, val)  # Insert into the left subtree
        elif val > root.val:
            root.right = self.insert(root.right, val)  # Insert into the right subtree
        
        # Return the unchanged root
        return root
```

```python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        curr = root

        while curr:
            # If both p and q are greater than the current node, move to the right subtree
            if p.val > curr.val and q.val > curr.val:
                curr = curr.right
            # If both p and q are less than the current node, move to the left subtree
            elif p.val < curr.val and q.val < curr.val:
                curr = curr.left
            # If one value is on the left and the other on the right (or one of them is equal to the current node), this is the LCA
            else:
                return curr
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isSymmetric(self, root: TreeNode) -> bool:
        if not root:
            return True  # An empty tree is symmetric
        
        def isMirror(t1: TreeNode, t2: TreeNode) -> bool:
            if not t1 and not t2:
                return True  # Both nodes are null, symmetric
            if not t1 or not t2:
                return False  # One is null, the other is not
            
            return (t1.val == t2.val and 
                    isMirror(t1.left, t2.right) and 
                    isMirror(t1.right, t2.left))
        
        return isMirror(root.left, root.right)
```

```python
from typing import List, Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        if not preorder or not inorder:
            return None  # Base case: empty tree
        
        # The first element in preorder is the root
        root_val = preorder[0]
        root = TreeNode(root_val)
        
        # Find the root in inorder to split left and right subtrees
        root_index_in_inorder = inorder.index(root_val)
        
        # Left subtree in preorder and inorder
        left_preorder = preorder[1:1 + root_index_in_inorder]
        left_inorder = inorder[:root_index_in_inorder]
        
        # Right subtree in preorder and inorder
        right_preorder = preorder[1 + root_index_in_inorder:]
        right_inorder = inorder[root_index_in_inorder + 1:]
        
        # Recursively build left and right subtrees
        root.left = self.buildTree(left_preorder, left_inorder)
        root.right = self.buildTree(right_preorder, right_inorder)
        
        return root
```
