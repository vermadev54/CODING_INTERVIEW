```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        start = 0  # Left boundary of the window
        max_length = 0  # Store the longest length found
        char_set = set()  # Track unique characters

        for end in range(len(s)):
            # If a duplicate character is found, shrink the window from the left
            while s[end] in char_set:
                char_set.remove(s[start])
                start += 1
            
            # Add the current character to the set
            char_set.add(s[end])

            # Update the maximum length found
            max_length = max(max_length, end - start + 1)
        
        return max_length
```

```python
s = "abcabcbb"
Solution().lengthOfLongestSubstring(s)
```

```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        left = 0  # Left boundary of the sliding window
        char_count = {}  # Dictionary to store character frequencies
        max_length = 0  # Store the longest valid substring length
        max_freq = 0  # Track the highest frequency of any character in the window
        
        for right in range(len(s)):
            # Increase frequency of the current character
            char_count[s[right]] = 1 + char_count.get(s[right], 0)
            
            # Update max_freq to avoid recalculating max(char_count.values())
            max_freq = max(max_freq, char_count[s[right]])

            # Check if window size is invalid (more than k replacements needed)
            while (right - left + 1) - max_freq > k:
                char_count[s[left]] -= 1
                left += 1  # Shrink window
            
            # Update maximum valid window size
            max_length = max(max_length, right - left + 1)

        return max_length
```

```python
s = "abcabcbb"
Solution().characterReplacement(s,2)
```

```python
from collections import defaultdict
from typing import List

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        anagram_groups = defaultdict(list)  # Map char frequency tuple -> list of words
        
        for word in strs:
            char_count = [0] * 26  # Frequency count for 'a' to 'z'
            
            for char in word:
                char_count[ord(char) - ord('a')] += 1  # Update count of each character
            
            # Convert list to tuple (hashable) and use it as a key
            anagram_groups[tuple(char_count)].append(word)
        
        return list(anagram_groups.values())  # Convert dict_values to list
```

```python
Solution().groupAnagrams(["eat","tea","tan","ate","nat","bat"])
```
